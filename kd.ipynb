```python
import pandas as pd

df=pd.read_csv('kidney_disease.csv')
df
```

```python
df.drop(columns="pcc")
df.drop(columns="pc")
```

```python
df.drop(columns="id")
```

```python
df=df.drop(columns=["id","pcc","pc","pot","htn","dm"])
```

```python
list(df.columns)
```

```python
df.head()
```

```python
df=df.dropna()
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
```

```python
display(df.dtypes)
```

```python
df.describe()
```

```python
df["bp"]=df["bp"].astype(int)
display(df.dtypes)
```

```python
df=df.astype({"age":'int',"bp":'int',"sg":'int'})
display(df.dtypes)
```

```python
df=df.astype({"al":'int',"su":'int',"bgr":'int',"bu":'int',"sc":'int',"sod":'int',"hemo":'int'})
display(df.dtypes)
```

```python
df['rbc']=l.fit_transform(df['rbc'])
df
```

```python
df["ba"]=l.fit_transform(df["ba"])
df["pcv"]=l.fit_transform(df["pcv"])
df["wc"]=l.fit_transform(df["wc"])
df["rc"]=l.fit_transform(df["rc"])
df["cad"]=l.fit_transform(df["cad"])
df["appet"]=l.fit_transform(df["appet"])
df["pe"]=l.fit_transform(df["pe"])
df["ane"]=l.fit_transform(df["ane"])
df["classification"]=l.fit_transform(df["classification"])
df
```

```python
df.rename(columns={"classification":"kidneydisease"},inplace=True)
df
```

```python
df.reset_index(drop=True,inplace=True)
df
```

```python
x=df.drop(columns="kidneydisease")
x
```
